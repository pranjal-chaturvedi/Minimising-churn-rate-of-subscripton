# Cleaning the data

```python
df.isna().sum()
```

Considerations: From the above code, we observe that 'credit_score' column 8031 rows which are NaN, 'rewards_earned' has 3227 rows as NaN and 'age' has 4 rows as NaN. So, we will get rid of the columns 'credit_score' and 'rewards_earned' as we cannot make an effective machine learning model even if we substitute Nulls with some metric, say mean. There is a caveat that can be discovered from this plot: The column 'app_web_user' is a function of the columns 'app_downloaded' and 'web_user'. If the person has downloaded the app AND also used the website, then the 'app_web_user' will be 1. It will be zero otherwise. This would therefore not be beneficial to include in the model.

Therefore, we can drop the 'app_web_user' from our DataFrame.

```python
df = df.drop("app_web_user", axis=1)
```

```python
# Removing the 'user' column as it is unique to each row and is not significant for out model
used_ID = df["user"]
df = df.drop("user", axis=1)
```

# One-hot Encoding

```python
df = pd.get_dummies(df, drop_first=True)
```

# Train/Test split and Feature Scaling

```python
X = df.drop("churn", axis=1)
```

```python
y = df["churn"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

# Training the model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(class_weight="balanced")
```

```python
model.fit(scaled_X_train, y_train)
```

```python
predictions = model.predict(scaled_X_test)
```

# Evaluation

```python
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
plot_confusion_matrix(model, scaled_X_test, y_test)
```

```python
print(classification_report(y_test, predictions))
```

From this, we can comclude that the model's accuracy is around 71% with similar values for Precision, Recall and F1-score. Further calculating the Cross Validation score.

```python
from sklearn.model_selection import cross_val_score
```

```python
accuracies = cross_val_score(estimator=model, X=scaled_X_train, y=y_train, cv=10)
```

```python
accuracies.mean()
```

Further analysing which features were more prominent as per the model.

```python
importances = pd.DataFrame(model.feature_importances_, columns=["feature importance score"])
```

```python
importances = pd.concat([importances, pd.DataFrame(X_train.columns, columns=["feature name"])], axis=1)
```

```python
plt.figure(figsize=(15, 6), dpi=200)
sns.barplot(data=importances.sort_values("feature importance score", ascending=False), x="feature name", y="feature importance score")
plt.title("Feature importance score for every feature variable in the data set")
plt.xticks(rotation=90);
```

# Conclusion
* Our model has provided us with an indication of which users are likely to churn. We have purposefully left the date of the expected churn open-ended because we are focussed on only gauging the features that indicate disengagement with the product, and not the exact manner (like timeframe) in which users will disengage. In this case study we have chosen this emphasis to get a sense of those who are just a _bit likely_ to churn because we are not aiming to create new products for people who are going to leave us for sure, but for people who are starting to lose interest in the app.

* If, after creating new product features, we start seeing our model predict that less of our users are going to churn, then we can assume our customers are feeling more engaged with what we are offering them. We can move forward with these efforts by enquiring the opinions of our users about our new features (e.g., polls, surveys, feedbacks etc.). If we want to transition into predicting churn more accurately, in order to put emphasis strictly on those leaving us, then we can add a time dimension to churn, which would add more accuracy to our model.
